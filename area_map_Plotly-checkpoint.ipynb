## Area Map (Choropleth Map)

#### An area map is used to visualize data across geographical regions.Different colors represent different values for each area.

## Import Libraries

```python
import pandas as pd
import plotly.express as px
```

 ## Example: Country-wise Population

```python
df2 = pd.DataFrame({
    "Country": ["IND", "USA", "CHN", "BRA", "RUS"],
    "GDP": [3.7, 26.9, 17.7, 4.2, 4.4]  # in trillion USD
})

fig = px.choropleth(
    df2,
    locations="Country",
    locationmode="ISO-3",
    color="GDP",
    title="Country-wise GDP Distribution",
    color_continuous_scale="Plasma"
)
fig.update_layout(
    width=1000,   # increase width
    height=600    # increase height
)
fig.write_image("area_map_example1.png")
fig.show()
```

## Example 2: Country-wise Internet Users

```python
df3 = pd.DataFrame({
    "Country": ["IND", "USA", "CHN", "BRA", "RUS"],
    "Internet_Users": [850, 310, 1050, 165, 215]  # in millions
})

fig = px.choropleth(
    df3,
    locations="Country",
    locationmode="ISO-3",
    color="Internet_Users",
    title="Country-wise Internet Users",
    color_continuous_scale="Viridis"
)
fig.update_layout(
    width=1000,   # increase width
    height=600    # increase height
)
fig.write_image("area_map_example2.png")
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Sample Data
df4 = pd.DataFrame({
    "Country": ["IND", "USA", "CHN", "BRA", "RUS"],
    "Population": [1420, 331, 1440, 213, 146]  # in millions
})

# Create Choropleth Map
fig = px.choropleth(
    df4,
    locations="Country",
    locationmode="ISO-3",
    color="Population",
    hover_name="Country",
    hover_data=["Population"],
    title="Country-wise Population Distribution",
    color_continuous_scale="Plasma"
)

# Customize Layout
fig.update_layout(
    width=1000,
    height=600
)

# Save Image
fig.write_image("population_map.png")

# Show Map
fig.show()
```
