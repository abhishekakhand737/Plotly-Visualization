## Scatter Plot

##### A scatter plot is used to show the relationship between two numerical variables.
##### Each point represents an observation in the dataset.

### Import Libraries

```python
import pandas as pd 
import plotly.express as px
```

### Example 1: Basic Scatter Plot

```python
df=pd.DataFrame({'Study_hr':[2,5,6,7,8,9],
                "Marks":[60,70,79,84,88,97]})
fig=px.scatter(df,x='Study_hr',y='Marks',title='Study Hours vs Marks')
fig.show()
```

## Example 2: Scatter Plot with Color

```python
fig=px.scatter(df,x='Study_hr',y='Marks',
               color='Marks',
               title='Scatter plot with color Ecoding')
fig.show()
```

## Example 3: Scatter Plot with Size

```python
fig=px.scatter(df,x='Study_hr',y='Marks',
               size='Marks',
               title='Size distribution using marks')
fig.show()
```

```python
fig = px.scatter(
    df,
    x='Hours',
    y='Marks',
    size='Attempts',
    color='Attempts',
    hover_data=['Hours', 'Marks', 'Attempts'],
    title='Enhanced Scatter Plot'
)

fig.show()
```

## Example 4: Scatter Plot with Hover Data

```python
fig=px.scatter(df,x='Study_hr',y='Marks',
               hover_data=['Study_hr','Marks'],
               title='Scatter plot with Hover information')
fig.show()
```

## Example 5: Categorical Scatter Plot

```python
df2 = pd.DataFrame({
    "Hours": [2, 4, 6, 3, 5, 7],
    "Score": [45, 70, 85, 60, 78, 92],
    "Result": ["Fail", "Pass", "Pass", "Pass", "Pass", "Pass"]
})

fig = px.scatter(
    df2,
    x="Hours",
    y="Score",
    color="Result",
    title="Student Performance by Result Category"
)
fig.show()
fig.write_image('Sactterplot.png')
```
