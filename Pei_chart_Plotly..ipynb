## Pie Chart

##### A pie chart in Plotly is used to represent proportional or percentage data as slices of a circle.

##### Each slice shows the contribution of a category to the whole dataset.

##### Plotly pie charts are interactive, allowing users to view values on hover.

##### They are best used when comparing a small number of categories.


### Import libraries

```python
import plotly.express as px
import pandas as pd 
```

## Example 1: Expense Distribution

```python
df = pd.DataFrame({
    "Category": ["Rent", "Food", "Transport", "Entertainment", "Savings"],
    "Amount": [15000, 8000, 3000, 2000, 5000]
})
```

```python
fig = px.pie(
    df,
    names="Category",
    values="Amount",
    title="Monthly Expense Distribution"
)
fig.show()
```

## Example 2: Student Result Share

```python
df2 = pd.DataFrame({
    "Result": ["Pass", "Fail"],
    "Count": [45, 12]
})
fig = px.pie(
    df2,
    names="Result",
    values="Count",
    title="Student Result Distribution"
)
fig.show()
```

## Example 3: Market Share

```python
df3 = pd.DataFrame({
    "Company": ["Brand A", "Brand B", "Brand C", "Brand D"],
    "Share": [40, 30, 20, 10]
})

fig = px.pie(
    df3,
    names="Company",
    values="Share",
    title="Market Share of Companies"
)
fig.show()
```

## Example 4: Mobile OS Market Share.

```python
df4 = pd.DataFrame({
    "Operating System": ["Android", "iOS", "Others"],
    "Users": [72, 26, 20]
})

fig = px.pie(
    df4,
    names="Operating System",
    values="Users",
    title="Mobile Operating System Market Share"
)
fig.show()
```

## Example 5: Time Spent in a Day

```python
df5 = pd.DataFrame({
    "Activity": ["Study", "Sleep", "Entertainment", "Travel", "Others"],
    "Hours": [6, 8, 4, 2, 4]
})

fig = px.pie(
    df5,
    names="Activity",
    values="Hours",
    title="Daily Time Distribution of a Student"
)
fig.write_image('Piechart.png')
fig.show()
```
